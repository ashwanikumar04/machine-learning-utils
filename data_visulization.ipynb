```python
import pandas as pd
import numpy as np
```

```python
car_loans = pd.read_csv('data/car_financing.csv')
car_loans.head(1)
```

```python
car_loans['car_type'].value_counts()
```

```python
car_loans = car_loans.rename(columns={'Month':'month','Starting Balance':'starting_balance','Repayment':'repayment'
                                      ,'Interest Paid':'interest_paid','Principal Paid':'principal_paid','New Balance':'new_balance' })
```

```python
car_loans.head(1)
```

```python
car_loans = car_loans.drop(columns=['term','repayment'])
```

```python
car_loans.info()
```

```python
car_loans.isna().sum()
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
car_loans = car_loans.loc[:59]
```

```python
car_loans.info()
```

```python
month_number = car_loans.loc[:,'month'].values
interest_paid = car_loans.loc[:,'interest_paid'].values
principal_paid = car_loans.loc[:,'principal_paid'].values
```

```python
plt.plot(month_number,interest_paid)
plt.plot(month_number,principal_paid)
```

```python
plt.style.available
```

```python
plt.style.use('classic')
plt.plot(month_number,interest_paid)
plt.plot(month_number,principal_paid)
```

```python
plt.style.use('seaborn')
plt.plot(month_number,interest_paid)
plt.plot(month_number,principal_paid)
```

# Matlab Style

```python
plt.style.use('seaborn')
plt.plot(month_number,interest_paid,c='k')
plt.plot(month_number,principal_paid,c='b')
plt.xlabel('Month',fontsize=15)
plt.ylabel('Dollars',fontsize=15)
plt.title('Principal and interest paid per month', fontsize=22)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.tight_layout()
```

# Object oriented Style

```python
fig, axes = plt.subplots(nrows=1, ncols=1)
axes.plot(month_number, interest_paid, c='k')
axes.plot(month_number, principal_paid, c='b')
axes.set_xlabel('Month',fontsize=15)
axes.set_ylabel('Dollars',fontsize=15)
axes.set_title('Principal and interest paid per month',fontsize=22)
axes.tick_params(axis='x',labelsize=15)
axes.tick_params(axis='y',labelsize=15)
fig.tight_layout()
```

# Combination

```python
fig, axes = plt.subplots(nrows=1, ncols=1)
plt.plot(month_number, interest_paid, c='k')
axes.plot(month_number, principal_paid, c='b')
```
